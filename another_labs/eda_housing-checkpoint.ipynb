# График распределения числовых данных (одномерное и парное совместное распределение)

 Cначала посмотрим на распределение цен

```python
sns.distplot(data.price);
```

Можем заметить,что большинство цен на аренду жилья меньше тысячи,и есть небольшое количество очень дорогого жилья,поэтому посмотрим отдельно на жилье с малой ценой 

```python
sns.distplot(data[data['price'] < 600].price);
```

Можем заметить,что больше всего жилья с ценой до 100 долларов

Также посмотрим на количество отзывов

```python
sns.distplot(data.number_of_reviews);
```

Тут также есть небольшое количество записей с большим количеством отзывов,но у большинства количество отзывово меньше 100.
Посмотрим подробнее

```python
sns.distplot(data[data.number_of_reviews < 20].number_of_reviews);
```

У большинства записей вообще нет отзывов, затем с увеличением числа отзывов уменьшается число записей.

Аналогичная ситуация будет с отзывами в месяц.

```python
sns.distplot(data[data.reviews_per_month.notna()].reviews_per_month);
```

Теперь посмотрим на график и гистограмму распределения доступности жилья.
Большинство доступно только очень малое количество дней в году.
Также есть большое количество,доступных 365 дней в году
Остальные записи примерно равномерно распределены между этими двумя значениями.

```python
sns.distplot(data.availability_365);
```

Также можем посмотреть на график суммарного количества сдаваемого жилья арендодателями.Для этого нужно сделать группировку по host_id

```python
data2=data.groupby('host_id')['calculated_host_listings_count'].count()
```

```python
sns.distplot(data2)
```

У большинства арендодателей мало сдаваемого жилья,но есть несколько крупных арендодателей

```python
data2[data2 > 50]
```

Теперь посмотрим на взаимосвязи между численными величинами

```python
sns.pairplot(data[['price', 'number_of_reviews', 'availability_365']], diag_kind = "kde")
```

Больше всего отзывов на дешевое жилье.
