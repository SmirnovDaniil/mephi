## 1) $P_f = 0.4$ $P_s = 0.2$ $P_c = 0.4$

```python
import random
import numpy as np
u = 1000000
u0 = 1000000# –∏—Å—Ö–æ–¥–Ω–æ–µ —á–∏—Å–ª–æ –¥–µ–ª—è—â–∏—Ö—Å—è —è–¥–µ—Ä
n = 1 # –∏—Å—Ö–æ–¥–Ω–æ–µ —á–∏—Å–ª–æ —Å–≤–æ–±–æ–¥–Ω—ã—Ö –Ω–µ–π—Ç—Ä–æ–Ω–æ–≤
# —á–∏—Å–ª–æ ùúà (–≥—Ä–µ—á–µ—Å–∫–∞—è ¬´–Ω—é¬ª) ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–π—Ç—Ä–æ–Ω–æ–≤,
# –∫–æ—Ç–æ—Ä—ã–µ –∏—Å–ø—É—Å–∫–∞—é—Ç—Å—è –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ —Ä–µ–∞–∫—Ü–∏–∏ –¥–µ–ª–µ–Ω–∏—è
p1 = 0.4 #–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–µ–ª–µ–Ω–∏—è(–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–æ–π–¥–µ—Ç –µ—Å–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–µ —á–∏—Å–ª–æ –æ–∫–∞–∂–µ—Ç—Å—è –≤ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ –æ—Ç 0 –¥–æ 0.4)
p2 = 0.6 #–ï—Å–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–µ —á–∏—Å–ª–æ –æ–∫–∞–∂–µ—Ç—Å—è –≤ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ –æ—Ç 0.4 –¥–æ 0.6 –ø—Ä–æ–∏–∑–æ–π–¥–µ—Ç —Ä–∞—Å—Å–µ—è–Ω–∏–µ,–≤ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Å–ª—É—á–∞—è—Ö –∑–∞—Ö–≤–∞—Ç
nu = 2
number_of_neutrons = [1]
number_of_cores = [1000000]
k_values = []
while u > 0: 
    n += 1
    n0 = n
    for i in range(n):
        p = random.random()
        if u > 0:
            if p < p1:
                n += (nu-1)
                u -= 1
            if p1 < p < p2:
                u = u
                n = n
            if p > p2:
                n -= 1
                u -= 1
    k = n / n0
    k_values.append(k)
    number_of_neutrons.append(n)
    number_of_cores.append(u)
print("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫–æ–ª–µ–Ω–∏–π –Ω–µ–π—Ç—Ä–æ–Ω–æ–≤, –∑–∞ –∫–æ—Ç–æ—Ä–æ–µ –∑–∞–∫–æ–Ω—á–∏–ª–∏—Å—å —è–¥—Ä–∞ —É—Ä–∞–Ω–∞ =", len(k_values))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline

fig, axs = plt.subplots(3, 1, figsize=(14, 20))


axs[0].plot([i for i in range(1, len(k_values) + 2)], number_of_neutrons)
axs[0].set_ylabel('–ß–∏—Å–ª–æ –Ω–µ–π—Ç—Ä–æ–Ω–æ–≤')
axs[0].set_xlabel('–ù–æ–º–µ—Ä –ø–æ–∫–æ–ª–µ–Ω–∏—è')
axs[0].xaxis.set_major_locator(ticker.MultipleLocator(100))
axs[0].yaxis.set_major_locator(ticker.MultipleLocator(100))
axs[0].grid()
axs[0].set_title('–ì—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏  —Ç–µ–∫—É—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–≤–æ–±–æ–¥–Ω—ã—Ö –Ω–µ–π—Ç—Ä–æ–Ω–æ–≤ –æ—Ç –Ω–æ–º–µ—Ä–∞ –ø–æ–∫–æ–ª–µ–Ω–∏—è', fontsize = 13,
                  color = 'red')


axs[1].plot([i for i in range(1, len(k_values) + 2)],number_of_cores)
axs[1].set_ylabel('–ß–∏—Å–ª–æ —è–¥–µ—Ä —É—Ä–∞–Ω–∞')
axs[1].set_xlabel('–ù–æ–º–µ—Ä –ø–æ–∫–æ–ª–µ–Ω–∏—è')
axs[1].xaxis.set_major_locator(ticker.MultipleLocator(100))
axs[1].yaxis.set_major_locator(ticker.MultipleLocator(50000))
axs[1].grid()
axs[1].set_title('–ì—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è —è–¥–µ—Ä —É—Ä–∞–Ω–∞ –æ—Ç –Ω–æ–º–µ—Ä–∞ –ø–æ–∫–æ–ª–µ–Ω–∏—è', fontsize = 13,
                  color = 'red')

axs[2].plot([i for i in range(1, len(k_values) + 1)], k_values)
axs[2].set_ylabel('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —Ä–∞–∑–º–Ω–æ–∂–µ–Ω–∏—è')
axs[2].xaxis.set_major_locator(ticker.MultipleLocator(100))
axs[2].yaxis.set_major_locator(ticker.MultipleLocator(0.2))
axs[2].grid()
axs[2].set_title('–ì—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ —Ä–∞–∑–º–Ω–æ–∂–µ–Ω–∏—è –Ω–µ–π—Ç—Ä–æ–Ω–æ–≤ –æ—Ç –Ω–æ–º–µ—Ä–∞ –ø–æ–∫–æ–ª–µ–Ω–∏—è', fontsize = 13,
                  color = 'red')


plt.xlabel('–ù–æ–º–µ—Ä –ø–æ–∫–æ–ª–µ–Ω–∏—è')

plt.show()
```

2)–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —Ä–∞–∑–º–Ω–æ–∂–µ–Ω–∏—è $k=\frac{n_0 + \Delta n}{n_0} = 1 + P_f (\nu-1) - P_c + \frac{1}{n_0} $
